```python
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
```

```python
plt.figure(figsize=(12, 8))

# Create a Pitch object
pitch = Pitch(pitch_type='statsbomb', line_color='white', line_zorder=2,
              pitch_color='grass', stripe=True)


pitch.draw()

# Player positions (x, y, player_name)
player_positions = [
    (7, 40, 'Ramsey'),  # Goalkeeper
    (30, 70, 'Moy'),  # Right-back
    (30, 5, 'Battle'),  # Left-back 
    (30, 50, 'Blundell'), # Right-center-back 
    (30, 25, 'Eriksson'),  # Left-center-back
    (50, 40, 'Dali'),  # Central-midfielder 
    (100, 70, 'Galton'), # Right-forward 
    (80, 50, 'Zelem'),   # Right-midfielder
    (110, 40, 'Kerr'),  # Center-forward
    (80, 25, 'Reiten'),  # Left-midfielder
    (100, 5, 'Daly')  # Left-forward
]

# Plot player positions
for position in player_positions:
    x, y, player_name = position
    plt.scatter(x, y, color='red', edgecolors='black', zorder=3)
    plt.text(x, y + 2.5, player_name, color='white', fontsize=12, ha='center', va='center')
    
# Add title
plt.title('2022/2023 Team of the Season')

# Save the plot
plt.savefig('2223TOTS.png')

plt.show()
```
